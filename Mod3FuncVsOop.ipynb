```python
def sort_array(arr):
    count0 = arr.count(0)
    count1 = arr.count(1)
    count2 = arr.count(2)
    return [0] * count0 + [1] * count1 + [2] * count2

arr = [1, 2, 2, 0, 1, 0] #example array

print(sort_array(arr))
```

```python
def binary_search(arr, k):
    low, high = 0, len(arr) - 1
    result = -1 #default

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == k:
            result = mid
            high = mid - 1
        elif arr[mid] < k:
            low = mid + 1
        else:
            high = mid - 1
        return result

arr = [1, 2, 3, 4, 5]
print(binary_search(arr, 4))
```
