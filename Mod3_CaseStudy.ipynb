```python
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

class Automobile(Vehicle):
    def __init__(self, vehicle_type, year, make, model, num_doors, roof_type):
        super().__init__(vehicle_type)
        self.year = year
        self.make = make
        self.model = model
        self.num_doors = num_doors
        self.roof_type = roof_type

    def print_info(self):
        print(f"Vehicle type: {self.vehicle_type}")
        print(f"Year: {self.year}")
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Number of Doors: {self.num_doors}")
        print(f"Type of Roof: {self.roof_type}")

def main():
    vehicle_type = "car"

    year = input("Please enter year (ex. 2018): ")
    make = input("Please enter make (ex. Chevrolet): ")
    model = input("Please enter model (ex. Spark): ")
    num_doors = input("Please enter number of doors (ex. 4): ")
    roof_type = input ("Please enter roof type (ex. solid or sun roof): ")

    car = Automobile(vehicle_type, year, make, model, num_doors, roof_type)

    car.print_info()

if __name__ == "__main__":
    main()
```
